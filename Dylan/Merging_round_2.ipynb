```python
# changing the ".." values to pd.NA, then making them numeric to match formatting
cleaned_world = cleaned_world.replace('..', pd.NA)
numeric_columns = cleaned_world.columns.drop(['Country Name', 'Country Code', 'Series Name', 'Series Code'])
cleaned_world[numeric_columns] = cleaned_world[numeric_columns].apply(pd.to_numeric, errors='coerce')
print(cleaned_world.shape)
cleaned_world.head(2)
```

```python
# match Country Code to NOC for merging
cleaned_world = cleaned_world.rename(columns = {"Country Code": "NOC"})
cleaned_world.head(2)
```
