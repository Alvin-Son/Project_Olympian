```python
# birth date to year 
current_year = datetime.now().year
df3["birth_date"] = pd.to_datetime(df3["birth_date"], format="%Y-%m-%d",errors="coerce")
df3["Age"] = current_year - df3["birth_date"].dt.year
df3.head()
```

```python
# remove 0 from height and weight
df3["height"].replace(0, np.nan, inplace=True)
df3["weight"].replace(0, np.nan, inplace=True)
df3.head(1)
```

```python
x =df3["country_code"].unique()
len(x)
```

```python
# total athletes
athlete_count = df3.groupby(["country_code", "country"])["name"].nunique().reset_index(name="Total Athletes")
print(athlete_count.shape)
athlete_count.head(1)
```

```python
df3 = df3.drop(columns = ["name", "function", "birth_date"])
```

```python
# aggregate age, height, weight by sex for each year and country
agg3 = df3.copy()
results_sex = []
start_time = time.time()

NOCs = agg3["country_code"].unique()
country_names = agg3["country"].unique()
print(len(NOCs))
print(len(country_names))
```

```python
# aggregate age, height, weight by sex for each year and country
agg3 = df3.copy()
results_sex = []
start_time = time.time()

NOCs = agg3["country_code"].unique()
country_names = agg3["country"].unique()

for N in NOCs:


    males = agg3[(agg3["country_code"] == N) & (agg3["gender"] == "Male")]
    females = agg3[(agg3["country_code"] == N) & (agg3["gender"] == "Female")]

    avg_age_m = males["Age"].mean()
    avg_height_m = males["height"].mean()
    avg_weight_m = males["weight"].mean()

    avg_age_f = females["Age"].mean()
    avg_height_f = females["height"].mean()
    avg_weight_f = females["weight"].mean()

    results_sex.append({
        "NOC": N,
        "Average Age (M)": avg_age_m,
        "Average Height (M)": avg_height_m,
        "Average Weight (M)": avg_weight_m,
        "Average Age (F)": avg_age_f,
        "Average Height (F)": avg_height_f,
        "Average Weight (F)": avg_weight_f
    })

athletes_df = pd.DataFrame(results_sex)
total_time = time.time() - start_time
minutes = total_time/60
print(f" This took {total_time} seconds.... or {minutes} minutes....")
athletes_df.head(50)
```
